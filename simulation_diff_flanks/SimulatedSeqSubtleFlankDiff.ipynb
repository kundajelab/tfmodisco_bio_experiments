```
import simdna
from simdna import synthetic as sn
loadedMotifs = sn.LoadedEncodeMotifs(simdna.ENCODE_MOTIFS_PATH, pseudocountProb=0.001)

embedders = []
for motifname, motifs, frequency in [
    ("GATA", (motifGATA_v1, motifGATA_v2), 0.5),
    ("TAL1", (motifTAL1_v1, motifTAL1_v2), 0.5),
    ("ELF1", (motifELF1_v1, motifELF1_v2), 0.5), ]:
  embedders.append(
      sn.RandomSubsetOfEmbedders(
          embedders=[
            sn.RepeatedEmbedder(
              embedder=sn.SubstringEmbedder(
                sn.ReverseComplementWrapper(
                    substringGenerator=sn.PwmSampler(motif),
                    reverseComplementProb=0.5
                ),
                positionGenerator=sn.UniformPositionGenerator(),
                name=motifname),
            quantityGenerator=sn.ZeroInflater(
                quantityGenerator=sn.FixedQuantityGenerator(1),
                zeroProb=1-frequency
            ))
        for motif in motifs
      ],
      quantityGenerator=sn.FixedQuantityGenerator(1))
  )

embedInBackground = sn.EmbedInABackground(
backgroundGenerator=sn.ZeroOrderBackgroundGenerator(seqLength=200),
embedders=embedders)

label_generator = sn.core.IsInTraceLabelGenerator(["GATA", "TAL1", "ELF1"])
#label_generator = sn.core.IsInTraceLabelGenerator(["TAL1"])

def labeling_func(self, generated_seq):
  labels = label_generator.generateLabels(generated_seq)
  labels.append(1 if np.sum(labels) >= 1 else 0)
  return labels

simdna.random.seed(1234)

sequenceSet = sn.GenerateSequenceNTimes(embedInBackground, 10000)
sn.printSequences("mysim.simdata", sequenceSet,
                          includeFasta=True, includeEmbeddings=True,
                          prefix="sim",
                          labelGenerator=label_generator
                          #labelGenerator=sn.core.LabelGenerator(label_generator.labelNames+["AnyTwo"],
                          #                                      labeling_func)
                          )
```

```
import numpy as np
import sklearn.model_selection

ltr = {
    'A': [1,0,0,0], 'C': [0,1,0,0], 'G': [0,0,1,0], 'T': [0,0,0,1]
}
def onehot_encode(seqs):
  return np.array([ [ltr[x] for x in seq] for seq in seqs])

all_labels = np.array([[float(y) for y in x.split("\t")[3:]] for i,x in
                    enumerate(open("mysim.simdata")) if i > 0])
sequences = [x.split("\t")[1] for i,x in enumerate(open("mysim.simdata"))
             if i > 0]
all_onehot = onehot_encode(sequences)

X_train, X_test, y_train, y_test, train_indices, test_indices = sklearn.model_selection.train_test_split(
    all_onehot, all_labels, np.arange(len(all_onehot)), test_size=0.2, random_state=1234)

!head -50 mysim.simdata
```

```
%tensorflow_version 1.x
import tensorflow as tf
import numpy as np
import keras
from sklearn.metrics import roc_auc_score


def set_seed(seed):
  np.random.seed(seed)
  tf.set_random_seed(seed)


def make_model(filter_width, filt_per_layer, num_layers):
  model = keras.models.Sequential()
  model.add(keras.layers.Conv1D(filters=filt_per_layer,
                                kernel_size=filter_width,
                                input_shape=X_train.shape[1:],
                                activation="relu",
                                kernel_initializer="he_normal"))
  for i in range(1,num_layers):
    model.add(keras.layers.Conv1D(filters=filt_per_layer,
                                kernel_size=filter_width,
                                activation="relu",
                                kernel_initializer="he_normal"))
  model.add(keras.layers.GlobalAveragePooling1D())
  #model.add(keras.layers.Dense(30, activation="relu"))
  #model.add(keras.layers.Dense(30, activation="relu"))
  model.add(keras.layers.Dense(3, activation="sigmoid"))
  return model


def train_model(modelcreator, seed):
  set_seed(seed)
  model = modelcreator()
  early_stopping_callback = keras.callbacks.EarlyStopping(
                              patience=20, restore_best_weights=True)
  model.compile(optimizer="adam", loss="binary_crossentropy",
                metrics=["accuracy"])
  history = model.fit(X_train, y_train,
            validation_data=(X_test, y_test), epochs=100,
            callbacks=[early_stopping_callback])
  model.set_weights(early_stopping_callback.best_weights)
  y_test_preds = model.predict(X_test)
  print(roc_auc_score(y_true=y_test, y_score=y_test_preds))
  return model

modarch1_factory = lambda: make_model(
    filter_width=7, filt_per_layer=30, num_layers=5)
print("Arch 1")
model = train_model(modarch1_factory, 100)
```

```
import tensorflow as tf
import deeplift.util
from deeplift.conversion import kerasapi_conversion as kc
from deeplift.dinuc_shuffle import dinuc_shuffle

model.save("model.h5")

deeplift_model =\
    kc.convert_model_from_saved_files(
        "model.h5",
        nonlinear_mxts_mode=deeplift.layers.NonlinearMxtsMode.Rescale) 
find_scores_layer_idx = 0
deeplift_hypcontribs_func = deeplift_model.get_target_multipliers_func(
                            find_scores_layer_idx=find_scores_layer_idx,
                            target_layer_idx=-2)

deeplift_shuffref_hypcontribs_func = deeplift.util.get_shuffle_seq_ref_function(
    score_computation_function=deeplift_hypcontribs_func,
    shuffle_func=dinuc_shuffle)

deeplifthypscores = np.array(deeplift_shuffref_hypcontribs_func(task_idx=[0,1,2],
                                         input_data_sequences=all_onehot,
                                         num_refs_per_seq=20,
                                         batch_size=256,
                                         progress_update=1000))
```
